## first let's import the necessary libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import openpyxl
```

```python
# Load the input file
input_data = pd.read_excel(r"C:\Users\Eyosi\Downloads\Input.xlsx")
```

```python
# Create a dictionary to store the extracted articles
articles = {}

for index, row in input_data.iterrows():
    url_id = row['URL_ID']
    url = row['URL']

    # Send a request to the URL
    response = requests.get(url)

    # Parse the response text with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Extract the article title and text
title = soup.find('title').text
text = soup.find('p').text

# Save the article in the dictionary
articles[url_id] = {'title': title, 'text': text}
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
# Initialize the sentiment intensity analyzer
sia = SentimentIntensityAnalyzer()
```

```python
for url_id, article in articles.items():
    text = article['text']

    # Tokenize the text
    tokens = nltk.word_tokenize(text)

    # Compute the sentiment scores
    sentiment_scores = sia.polarity_scores(text)
```

```python
# Save the sentiment scores in the dictionary
articles[url_id]['sentiment_scores'] = sentiment_scores
```

```python
# Convert the dictionary to a DataFrame
output_data = pd.DataFrame.from_dict(articles, orient='index')

# Save the DataFrame to an Excel file
output_data.to_excel('output.xlsx')
```
